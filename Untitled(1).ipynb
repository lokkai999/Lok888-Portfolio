```python
# Select columns from index 4 to 11
columns_to_convert = df.columns[4:12] 

# Loop through the columns and convert them to a numeric type
for col in columns_to_convert:
    # errors='coerce' will turn any value that cannot be converted into 'NaN'
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Verify the changes
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
# Select only the numeric columns that have missing values
numeric_cols_with_na = df.select_dtypes(include=['float64', 'int64']).columns
cols_to_fill = [col for col in numeric_cols_with_na if df[col].isnull().any()]

# Fill NaN values in each column with that column's median
for col in cols_to_fill:
    median_val = df[col].median()
    df[col].fillna(median_val, inplace=True)

# Verify that all null values have been handled
print("Missing values after imputation:")
print(df.isnull().sum())
```
